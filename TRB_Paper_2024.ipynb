```python
import pandas as pd

df=pd.read_excel('filtered_data_los angles _zip.xlsx')


# Convert 'Model Year' column to integer (if it is stored as a string)
df['Model Year'] = pd.to_numeric(df['Model Year'], errors='coerce')

# Filter the dataset to only include Model Year 2018
df_2018 = df.query("`Model Year` == 2018")

# Group the filtered dataset by 'Make' and 'Fuel' columns and calculate the sum of 'Vehicles' in each group
vehicles_by_Fuel_Duty = df_2018.groupby(['Make', 'Fuel'])['Vehicles'].sum().sort_values(ascending=False)

# Print the results
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(vehicles_by_Fuel_Duty)

# Calculate the total number of vehicles in the filtered DataFrame for Model Year 2018
total_vehicles_2018 = df_2018['Vehicles'].sum()
print("Total number of vehicles in 2018:", total_vehicles_2018)
```

```python
# Convert the Series to a DataFrame
vehicles_by_Fuel_Duty_df = vehicles_by_Fuel_Duty.reset_index()

# Save the results as an Excel file
vehicles_by_Fuel_Duty_df.to_excel('vehicles_by_make_fuel_2018_zip.xlsx', index=False)

# Save the results as an CSV file
vehicles_by_Fuel_Duty_df.to_csv('vehicles_by_make_fuel_2018_zip.csv', index=False)
```

```python
#(b) Vehicle Numbers by Source Type 
#Bar Chart
import seaborn as sns
import matplotlib.pyplot as plt

# Data
source_type = ['Passenger Car (Gasoline)', 'Passenger Car (Electricity)', 'Passenger Car (Ethanol)',
               'Passenger Truck (Gasoline)', 'Light Commercial Truck (Diesel)', 'Passenger Truck (Diesel)',
               'Combination Long-haul Truck (Diesel)', 'Light Commercial Truck (Gasoline)',
               'Combination Long-haul Truck (CNG)', 'Passenger Car (Diesel)', 'Single Unit Short-haul Truck (Diesel)',
               'Passenger Truck (Ethanol)', 'Single Unit Short-haul Truck (CNG)', 'Light Commercial Truck (CNG)',
               'Transit Bus (CNG)', 'Refuse Truck (Diesel)', 'Other Buses (Diesel)']

vehicle_numbers = [346634, 20983, 4393, 4249, 1371, 728, 636, 421, 405, 364, 357, 252, 111, 102, 60, 25, 24]

# Plotting the bar chart with reversed blue shade
plt.figure(figsize=(10, 8))
sns.barplot(x=vehicle_numbers, y=source_type, palette='Blues_r')  # Use 'Blues_r' for the reversed blue shade palette
plt.xlabel('Number of Vehicles', fontsize=14)
plt.ylabel('Source Type Name / Make', fontsize=14)
plt.title('Vehicle Numbers by Source Type', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Adding annotations to the bars
for index, value in enumerate(vehicle_numbers):
    # For "Passenger Car (Gasoline)" bar, annotate with its actual value
    if source_type[index] == 'Passenger Car (Gasoline)':
        plt.text(value + 50, index, f"{value}", va='center', fontsize=10, color='black', fontweight='bold')
    else:
        plt.text(value + 50, index, f"{value}", va='center', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

```python
#(a) Vehicle Distribution by Fuel Type
#Pie Chart
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data
fuel_types = ['Gasoline', 'Ethanol (E-85)', 'Diesel Fuel', 'Electricity', 'Compressed Natural Gas (CNG)']
number_of_vehicles = [351304, 4645, 3505, 20983, 678]
df = pd.DataFrame({'Fuel': fuel_types, 'Number of Vehicles': number_of_vehicles})

# Create a figure and axes
fig, ax = plt.subplots(figsize=(8, 8))

# Use different blue color shades and reverse the order
colors = list(reversed(sns.color_palette('Blues', len(fuel_types))))

# Plot the pie chart with 15-degree rotation
wedges, texts, autotexts = ax.pie(df['Number of Vehicles'], autopct='%1.1f%%', startangle=15, colors=colors)

# Add a circle at the center to create the donut shape
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Add a title
ax.set_title('Vehicle Distribution by Fuel Type')

# Add a legend
ax.legend(wedges, df['Fuel'], loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Show the pie chart
plt.show()
```
